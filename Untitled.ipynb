```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew


# Step 2: Load the dataset
file_path = 'diabetes_binary_health_indicators_BRFSS2015.csv'  # Update the path if needed
data = pd.read_csv(file_path)

# Step 3: Preview the dataset
print("Dataset Preview:")
print(data.head())
print("\nDataset Information:")
data.info()
```

```python
# Step 1: Check for missing values
print("Missing Values Count:")
print(data.isnull().sum())

# Step 2: Summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Step 3: Detect outliers (using boxplot for BMI as an example)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(x=data['BMI'])
plt.title("Boxplot for BMI")
plt.show()
```

```python


# Visualize boxplots for all numerical columns
def plot_boxplots(data, columns, ncols=3):
    nrows = (len(columns) + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows * 5))
    axes = axes.flatten()

    for i, col in enumerate(columns):
        sns.boxplot(y=data[col], ax=axes[i])
        axes[i].set_title(f"Boxplot for {col}")
        axes[i].set_ylabel("")
        axes[i].grid(True)

    # Hide unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# List of numerical columns to visualize
numerical_columns = data.select_dtypes(include=[np.number]).columns

# Plot boxplots
plot_boxplots(data, numerical_columns)
```

```python


# Function to plot histograms and KDE plots for each column
def visualize_skewness(data, columns):
    ncols = 3
    nrows = (len(columns) + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows * 5))
    axes = axes.flatten()

    for i, col in enumerate(columns):
        sns.histplot(data[col], kde=True, ax=axes[i], color="blue", bins=30)
        skew_value = skew(data[col].dropna())  # Drop NaN for skew calculation
        axes[i].set_title(f"{col} (Skewness: {skew_value:.2f})")

    for j in range(i + 1, len(axes)):
        axes[j].axis('off')  # Hide unused subplots

    plt.tight_layout()
    plt.show()

# Visualize skewness for numerical columns
visualize_skewness(data, numerical_columns)
```

# --Handling Preprocessing

```python
# Function to handle outliers using IQR
def handle_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column.clip(lower=lower_bound, upper=upper_bound)

# Apply the IQR method to all numerical columns
for col in numerical_columns:
    data[col] = handle_outliers_iqr(data[col])

# Visualize boxplots again to confirm changes
plot_boxplots(data, numerical_columns)



# Function to apply transformations for skewed data
def transform_skewed_columns(data, columns, threshold=0.5):
    for col in columns:
        skew_value = skew(data[col].dropna())
        if abs(skew_value) > threshold:  # Only transform if skewness > threshold
            print(f"Transforming {col} (Skewness: {skew_value:.2f})")
            if skew_value > 0:  # Right-skewed
                data[col] = np.log1p(data[col])  # log1p handles log(0)
            else:  # Left-skewed
                data[col] = np.sqrt(data[col].max() - data[col] + 1)
    return data

# Apply transformations
data = transform_skewed_columns(data, numerical_columns)

# Visualize again after transformations
visualize_skewness(data, numerical_columns)
```

# Feature Selection and Class distribution analysis

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from collections import Counter

# Step 1: Define features and target
X = data[['HighBP', 'HighChol', 'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'DiffWalk', 'BMI', 'GenHlth', 'Age', 'Income']]
y = data['Diabetes_binary']

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_scaled, y)

# Step 4: Check the new class distribution
print("Original Class Distribution:", Counter(y))
print("Balanced Class Distribution:", Counter(y_balanced))

# Step 5: Split the balanced data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)

print("Training Set Shape:", X_train.shape)
print("Testing Set Shape:", X_test.shape)
```

# ML Models Training and evaluations

Logistic regression
----------------------

```python
  
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt

# Step 1: Train the Logistic Regression model
logistic_model = LogisticRegression(random_state=42, max_iter=1000)
logistic_model.fit(X_train, y_train)

# Step 2: Make predictions
y_pred = logistic_model.predict(X_test)
y_pred_proba = logistic_model.predict_proba(X_test)[:, 1]

# Step 3: Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[0, 1])

# Step 4: ROC and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Visualizations
plt.figure(figsize=(15, 5))

# Plot Confusion Matrix
plt.subplot(1, 2, 1)
disp.plot(cmap='Blues', ax=plt.gca(), colorbar=False)
plt.title("Confusion Matrix")

# Plot ROC Curve
plt.subplot(1, 2, 2)
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})", color="blue")
plt.plot([0, 1], [0, 1], linestyle='--', color='red', label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()

```

# Hyper parameter Tuning

```python


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import RandomizedSearchCV
import matplotlib.pyplot as plt

# Define a smaller parameter grid
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}

# Step 1: Initialize Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Step 2: Perform Randomized Search for Hyperparameter Tuning
random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid,
    n_iter=20,  # Reduced number of combinations for faster computation
    scoring='roc_auc',
    cv=3,  # Use 3-fold cross-validation
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

# Step 3: Best Parameters and ROC-AUC from Tuning
best_params = random_search.best_params_
best_score = random_search.best_score_
print("Best Parameters:", best_params)
print("Best ROC-AUC (CV):", best_score)

# Step 4: Train the Best Model on Full Training Data
best_rf_model = random_search.best_estimator_
best_rf_model.fit(X_train, y_train)

# Step 5: Make Predictions
y_pred = best_rf_model.predict(X_test)
y_pred_proba = best_rf_model.predict_proba(X_test)[:, 1]

# Step 6: Evaluate the Tuned Model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[0, 1])

# ROC and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Visualizations
plt.figure(figsize=(15, 5))

# Plot Confusion Matrix
plt.subplot(1, 2, 1)
disp.plot(cmap='Blues', ax=plt.gca(), colorbar=False)
plt.title("Confusion Matrix")

# Plot ROC Curve
plt.subplot(1, 2, 2)
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})", color="blue")
plt.plot([0, 1], [0, 1], linestyle='--', color='red', label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
